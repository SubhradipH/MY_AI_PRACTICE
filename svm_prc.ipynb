```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target']=iris.target
df.head()
```

```python
iris.target_names
```

Target
0=setosa
1=versicolor
2=virginica

```python
df[df.target==1].head()
```

```python
#Creating a new column
df['flower_name']=df.target.apply(lambda x:iris.target_names[x])
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['target','flower_name'], axis='columns')
x.head()
```

```python
y=df.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.svm import SVC
model=SVC()
# By changing C and gamma we can improve the accuracy
#model=SVC(C=10,gamma=10)(Example)
#we can also use kernal by changing the RBF to other like linear,poly and etc
# we can check the kernals avilable by hovering on the SVC() and it will open the help where we can se available kernals
# changing the kernal we can also imporve the accuracy   
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
