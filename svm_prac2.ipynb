```python
import pandas as pd
from sklearn.datasets import load_wine
d=load_wine()
```

```python
dir(d)
```

```python
d.feature_names
```

```python
df=pd.DataFrame(d.data,columns=d.feature_names)
df.head()
```

```python
df['target']=d.target
df.head()
```

```python
d.target_names
```

```python
df[df.target==1].head()
```

```python
df['wine_name']=df.target.apply(lambda x : d.target_names[x])
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
df0=df[df.target==0]
df1=df[df.target==1]
df2=df[df.target==2]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['target','wine_name'], axis='columns')
x.head()
```

```python
y=df.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.svm import SVC
model=SVC(C=10)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
